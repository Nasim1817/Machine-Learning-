```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Dataset Read
```

```python
data = pd.read_csv('/content/drive/MyDrive/Salary Data.csv')
#explore Dataset
print(data.head())
```

```python
#Select Features
x = data[['YearsExperience']]
y = data['Salary']
```

```python
#spliat data set in form of training///testing
train_x,test_x,train_y,test_y = train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
#create Model for training
model = LinearRegression()
model.fit(train_x, train_y)
```

```python
#Make a predicaton Test Data
y_pred = model.predict(test_x)
```

```python
mse = mean_squared_error(test_y, y_pred)
r2 = r2_score(test_y, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
import matplotlib.pyplot as plt

plt.scatter(test_y, y_pred)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs. Predicted Values')
plt.show()
```

```python
# Visualize the results
plt.scatter(test_x, test_y, color='black')
plt.plot(test_x, y_pred, color='blue', linewidth=3)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Linear Regression: Years of Experience vs. Salary')
plt.show()
```
